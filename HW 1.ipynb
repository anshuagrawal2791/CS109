```python
def get_poll_xml(poll_id):
    user_agent = 'Mozilla/5.0 (Windows NT 6.1; Win64; x64)'
    headers = {'User-Agent' : 'super happy flair bot by /u/spladug'}
    url  = "http://charts.realclearpolitics.com/charts/%i.xml" %int(poll_id)
    req = urllib2.Request(url, headers=headers)
    page = urllib2.urlopen(req).read()
    soup = BeautifulSoup(page)
    return soup
    
```

```python
def _strip(s):
    return re.sub(r'[\W_]+','',s)
```

```python
def plot_colors(xml):
    print(len(xml.find_all('graph')))
    result = {}
    for graph in xml.find_all('graph'):
        title = _strip(graph.get('title'))
#         print title
        result[title] = graph.get('color')
    return result  
    
```

```python
plot_colors(soup)
```

```python
def rcp_poll_data(xml):
    dates =[]
    graphs = {}
    for entry in xml.find_all('series')[0].find_all('value'):
        dates.append(entry.string) 
    result = pd.DataFrame({'date':pd.to_datetime(dates)})
    for graph in xml.find_all('graph'):
        result[_strip(graph.get('title'))]= pd.to_numeric([(i.string) for i in graph.find_all('value')])
    return result
```

```python
def poll_plot(poll_id):
    xml = get_poll_xml(poll_id)
#     xml = soup
    data = rcp_poll_data(xml)
    colors = plot_colors(xml)
#     data['Obama']=data['Obama']/(data.Obama.sum()*0.01)
#     data['Romney']=data['Romney']/(data.Romney.sum()*0.01)

#     print colors.keys()
#     print data.columns.values
    norm = data[colors.keys()].sum(axis=1) / 100 
#     print norm
    for c in colors.keys():
        data[c] /= norm
#     print data.head(len(data))
#     print data['Romney'].sum()
#     for c in %colors
    data = data.rename(columns = {c: _strip(c) for c in data.columns})
    for label, color in colors.items():
        plt.plot(data.date,data[label],color=color,label=label)
    plt.xticks(rotation=70)
    plt.legend(loc='best')
    plt.xlabel("Date")
    plt.ylabel("Normalized Poll Percentage")
    
```

```python
from fnmatch import fnmatch
```

```python
def find_governor_races(soup):
    urls = []
    pattern = '*/epolls/????/governor/??/*-*.html'
    for i in soup.find_all('a'):
        if(i.get('href')):
            if (fnmatch(i.get('href'),pattern)):
                urls.append(i.get('href'))
    pattern2 = 'http://www.realclearpolitics.com/*' 
    for i in range(len(urls)):
        if (not fnmatch(urls[i],pattern2)):
            urls[i] = 'http://www.realclearpolitics.com'+urls[i]
    return urls
```

```python
def race_result(url):
    user_agent = 'Mozilla/5.0 (Windows NT 6.1; Win64; x64)'
    headers = {'User-Agent' : user_agent}
    req = urllib2.Request(url, headers=headers)
    page = urllib2.urlopen(req).read()
    soup = BeautifulSoup(page)
#     print soup.find('table',class_='data').find_all('tr')[1].find_all('td')[4].string
    result = {}
    for i in [3,4,5]:
        if i==5:
            if soup.find('table',class_='data').find_all('tr')[0].find_all('th')[i].string!="Spread":
                result[_strip(str(soup.find('table',class_='data').find_all('tr')[0].find_all('th')[i].string.split()[0]))]=float(soup.find('table',class_='data').find_all('tr')[1].find_all('td')[i].string) 
        if i!=5:
            result[_strip(str(soup.find('table',class_='data').find_all('tr')[0].find_all('th')[i].string.split()[0]))]=float(soup.find('table',class_='data').find_all('tr')[1].find_all('td')[i].string)  
    return result
    
```

```python
def id_from_url(url):
    """Given a URL, look up the RCP identifier number"""
    return url.split('-')[-1].split('.html')[0]
```

```python
def plot_race(url):
    """Make a plot summarizing a senate race
    
    Overplots the actual race results as dashed horizontal lines
    """
    #hey, thanks again for these functions!
    print url
    id = id_from_url(url)
    print id
    xml = get_poll_xml(id) 
#     print xml.prettify()
    colors = plot_colors(xml)

    if len(colors) == 0:
        return
    
    #really, you shouldn't have
    result = race_result(url)
#     print result
    
    poll_plot(id)
    plt.xlabel("Date")
    plt.ylabel("Polling Percentage")
    for r in result:
        plt.axhline(result[r], color=colors[_strip(r)], alpha=0.6, ls='--')
```

```python
def party_from_color(color):
    if color in ['#0000CC', '#3B5998']:
        return 'democrat'
    if color in ['#FF0000', '#D30015']:
        return 'republican'
    return 'other'
```

```python
def error_data(url):
    id = id_from_url(url)
    xml = get_poll_xml(id)
    
    colors = plot_colors(xml)
    if len(colors) == 0:
        return pd.DataFrame()
    
    
    result = race_result(url)
#     print (result)
    df = rcp_poll_data(xml)
#     result = race_result(url)
    
    df = df.rename(columns={c: _strip(c) for c in df.columns})
    
    for k,v in result.items():
        result[_strip(k)]=v
    
    candidates = [c for c in result.keys()]
    
#     print df.head(10)
#     df.index = df.date
#     df = df.resample('D').mean()
    df.dropna(inplace = True)
#     print df.head(10)
    
    forecast_length= (df.date.max() - df.date).values
    forecast_length = forecast_length/np.timedelta64(1,'D')
    
    errors = {}
    normalized = {}
    poll_lead = {}
    
    for c in candidates:
        corr = df[c]/df[candidates].sum(axis=1)*100
        err = corr-result[c]
        normalized[c] = corr
        errors[c] = err
        
#     print 'normalised----',normalized
#     print 'errors---',errors
        
    result = {}
    n = forecast_length.size
    result['percentage'] = np.hstack(normalized[c] for c in candidates)
    result['errors'] = np.hstack(errors[c] for c in errors)
    result ['candidate'] = np.hstack(np.repeat(c,n) for c in candidates)
    result['party'] = np.hstack(np.repeat(party_from_color(colors[_strip(c)]), n) for c in candidates)
    result['forecast_length'] = np.hstack(forecast_length for _ in candidates)
#     print result
    result = pd.DataFrame(result)
    return result
```
