```python
def simulate_prizedoor(nsim):
    return np.random.randint(0,3,nsim)
```

```python
def simulate_guess(nsim):
    return np.zeros(nsim,dtype=np.int64)
```

```python
def goat_door(prizedoors, guesses):
    result = np.random.randint(0,3,prizedoors.size)
#     print (prizedoors)
#     print (guesses)
#     print(result)
    while True:
        bad = (result == prizedoors) | (result == guesses)
#         print (bad)
        if not bad.any():
            return result
#         print(bad.sum())
#         print(result[bad])
        result[bad] = np.random.randint(0, 3, bad.sum())
        
def switch_guess(guesses, goatdoors):
    result = np.zeros(guesses.size)
    switch = {(0, 1): 2, (0, 2): 1, (1, 0): 2, (1, 2): 1, (2, 0): 1, (2, 1): 0}
    for i in [0, 1, 2]:
        for j in [0, 1, 2]:
            mask = (guesses == i) & (goatdoors == j)
            
            if not mask.any():
                continue
            result = np.where(mask, np.ones_like(result) * switch[(i, j)], result)
    return result
def switch_guess2(guesses, goatdoors):
    result = np.zeros(guesses.size)
    switch = {(0, 1): 2, (0, 2): 1, (1, 0): 2, (1, 2): 1, (2, 0): 1, (2, 1): 0}
    for i in range (guesses.size):
        result[i] = switch[guesses[i],goatdoors[i]]
    return result

def win_percentage(guesses,prizedoors):
    
    return 100*(guesses==prizedoors).mean()
```

```python
nsim = 20000
prizedoors = simulate_prizedoor(nsim)
guesses = simulate_guess(nsim)
goat_doors = goat_door(prizedoors,guesses)
print(prizedoors)
print(guesses)
print(goat_doors)

# goat_door(prizedoors,guesses)
# switches = switch_guess(guesses,goat_doors)
switches2 = switch_guess2(guesses,goat_doors)
# print(switches)
print (switches2)

print(win_percentage(guesses,prizedoors))
print(win_percentage(switches2,prizedoors))

```
